```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

#load data of bids from all 17 models, named as agents
data = pd.read_csv('/content/gdrive/My Drive/Data/combined.csv')
```

```python
data.head()
```

```python
#Series of iterations needed to compare bids among different agents per every  bid request

Agent =[]
Clicks= []
Impressions=[] 
Spend =[] 
CTR =[] 
CPM =[] 
eCPC =[]
impression_won=[]


    
    

for i in range(1,18):
    budget = 6250000
    impressions = 0
    clicks = 0
    bidprices_for_agent = data['Agent_'+str(i)].values
    data_for_agents = data.drop(columns=['Agent_'+str(i)])
    data_for_agents = data_for_agents.drop(columns=['click'])
    
    for x, y in data_for_agents.iterrows():
      
        bidprice = bidprices_for_agent[x]
        payprice=y.max()
        entry1=-1
        entry2=-1
        
        if (bidprice>=payprice): 
          budget= budget-payprice

          if (budget>=0): 
            entry1= x
            impressions+=1
            if (data['click'][x]==1):
              clicks +=1
              entry2= x
            entry ={'Agent': i, 'Impression ID' : entry1, 'Click ID': entry2 }
            impression_won.append(entry)
          
          else:
            budget= budget+payprice
            break
            

    Spent = 6250000 - budget
    
    if (impressions >0):
        CtR = clicks/impressions
        cPM = 1000*Spent/impressions
    else:
        CtR =0
        cPM =0

    
   
    
    if (clicks >0):
        ecPC = Spent/clicks
    else:
        ecPC =0

    Agent.append(i)
    Clicks.append(clicks)
    Impressions.append(impressions)
    Spend.append(Spent)
    CTR.append(CtR)
    eCPC.append(ecPC)
    CPM.append(cPM)
    print ('For Agent =',i, ' impressions =',impressions,' Clicks =',clicks,' Spent amount = ',Spent, ' eCPC is ',ecPC) 
    print ( ' CTR =',CtR, ' CPM is ',cPM)
    print (' ------->')
        
```

```python
Agent_bid_info=pd.DataFrame(Agent, columns=['Agent_ID'])
Agent_bid_info['Number of Impressions']=Impressions
Agent_bid_info['Number of Clicks']=Clicks
Agent_bid_info['Spent Amount']=Spend
Agent_bid_info['CTR']=CTR
Agent_bid_info['CPC']=eCPC
Agent_bid_info['CPM']=CPM
```

```python
Agent_bid_info.to_csv(r'/content/gdrive/My Drive/Data/multi-agent_based_model_competition.csv')
```
