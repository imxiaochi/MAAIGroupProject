```python
import pandas as pd
import numpy as np
```

```python
data_all = pd.read_csv('/content/gdrive/My Drive/Multiagent/Data/split_data_all3.csv')
```

```python
data_all.head()
```

```python
one = pd.get_dummies(data_all['os'],prefix = 'os')
```

```python
data_all = data_all.join(one)
```

```python
data_all = data_all.drop(columns = ['Unnamed: 0', 'os'])
```

```python
data_all.head()
```

```python
one1 = pd.get_dummies(data_all['browser'],prefix = 'browser')
data_all = data_all.join(one1)
data_all = data_all.drop(columns= ['browser'])
```

```python
data_all.head()
```

```python
one1 = pd.get_dummies(data_all['slotsize'],prefix = 'slotsize')
data_all = data_all.join(one1)
data_all = data_all.drop(columns= ['slotsize'])
```

```python
data_all.head()
```

```python
data_all.to_csv('/content/gdrive/My Drive/Multiagent/Data/split_data_all.csv', sep = ',')
```
