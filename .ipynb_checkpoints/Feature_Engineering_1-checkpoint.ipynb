```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.utils import resample
```

```python
# df = pd.read_csv('/content/gdrive/My Drive/we_data/train.csv')
df = pd.read_csv('/Desktop/jupyternb/we_data/train.csv')
df.head()
```

```python
cat_vars=['region','weekday','useragent','adexchange','slotwidth','slotheight','slotvisibility','advertiser']

for i in cat_vars:
    one_hot = pd.get_dummies(df[i],prefix=i)
    df = df.drop(i,axis = 1)
    df = df.join(one_hot)

df.head()
```

```python
# Downsample majority class and combine minority class with downsampled majority class
df_majority = pd.concat([resample(df[df.click==0], replace=False, n_samples=10000, random_state=123), df[df.click==1]])
 
# Upsample minority class and Combine majority class with upsampled minority class
df1 = pd.concat([df_majority, resample(df[df.click==1], replace=True, n_samples=10000-1793, random_state=123)])
 
# Display new class counts
df1.head()
```
