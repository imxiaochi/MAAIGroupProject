# MERGING TRAINING SET

```
# Warning: May crash due to large dataset when u proceed further!
train1 = pd.read_csv('/content/gdrive/My Drive/we_data/FOR_MODELLING/train_hash2.csv')
train1.head()
```

```
train1 = train1.drop('Unnamed: 0', axis=1)
# print(list(test1))
```

```
rem = ['urlid','slotwidth', 'slotheight','usertag', 'useragent1', 'useragent2', 'useragent3', 
       'useragent4', 'useragent5', 'IP1', 'IP2', 'IP3', 'IP4', 'IP5', 'IP6', 'IP7', 'IP8',
       'IP9', 'IP10','url3', 'url4', 'url5', 'url6', 'url7', 'url8', 'url9', 'url10', 'slotid1', 'slotid2',
       'slotid3', 'slotid4', 'slotid5', 'slotid6', 'slotid7', 'slotid8', 'slotid9', 'slotid10']
train1 = train1.drop(rem, axis = 1)
train1.head()
```

```
train1['slotvisibility_FirstView'] = train1['slotvisibility_FirstView'] + train1['slotvisibility_1']
train1['slotvisibility_SecondView'] = train1['slotvisibility_SecondView'] + train1['slotvisibility_2']
train1['slotvisibility_Na'] = train1['slotvisibility_Na'] + train1['slotvisibility_0']
train1['slotvisibility_OtherView'] = train1['slotvisibility_OtherView'] + train1['slotvisibility_255']

slotvisibility = ['slotvisibility_FirstView','slotvisibility_SecondView','slotvisibility_Na','slotvisibility_OtherView']

for i in slotvisibility:
  for j in range(len(train1)):
    if train1[i].loc[j] > 1:
      train1[i].loc[j] = 1
```

```
train2 = pd.read_csv('/content/gdrive/My Drive/split_data/split_data_test.csv')
train2 = train2.drop(['Unnamed: 0','Unnamed: 0.1'], axis = 1)
train2['bidid'] = train2['bidid']
train2.head()
```

```
train_final = pd.merge(train1, train2, on='bidid')
train_final.head()
```

```
train_final.to_csv('/content/gdrive/My Drive/we_data/FINAL_CSVs/FINAL_TRAIN_v1')
```
