# MERGING TEST SET

```
test1 = pd.read_csv('/content/gdrive/My Drive/we_data/FOR_MODELLING/test_hash2.csv')
test1.head()
```

```
test1 = test1.drop('Unnamed: 0', axis=1)
# print(list(test1))
```

```
rem = ['urlid','slotwidth', 'slotheight','usertag', 'useragent1', 'useragent2', 'useragent3', 
       'useragent4', 'useragent5', 'IP1', 'IP2', 'IP3', 'IP4', 'IP5', 'IP6', 'IP7', 'IP8',
       'IP9', 'IP10','url3', 'url4', 'url5', 'url6', 'url7', 'url8', 'url9', 'url10', 'slotid1', 'slotid2',
       'slotid3', 'slotid4', 'slotid5', 'slotid6', 'slotid7', 'slotid8', 'slotid9', 'slotid10']
test1 = test1.drop(rem, axis = 1)
test1.head()
```

```
test1['slotvisibility_FirstView'] = test1['slotvisibility_FirstView'] + test1['slotvisibility_1']
test1['slotvisibility_SecondView'] = test1['slotvisibility_SecondView'] + test1['slotvisibility_2']
test1['slotvisibility_Na'] = test1['slotvisibility_Na'] + test1['slotvisibility_0']
test1['slotvisibility_OtherView'] = test1['slotvisibility_OtherView'] + test1['slotvisibility_255']

slotvisibility = ['slotvisibility_FirstView','slotvisibility_SecondView','slotvisibility_Na','slotvisibility_OtherView']

for i in slotvisibility:
  for j in range(len(test1)):
    if test1[i].loc[j] > 1:
      test1[i].loc[j] = 1
```

```
test2 = pd.read_csv('/content/gdrive/My Drive/split_data/split_data_test.csv')
test2 = test2.drop(['Unnamed: 0','Unnamed: 0.1'], axis = 1)
test2['bidid'] = test1['bidid']
test2.head()
```

```
test_final = pd.merge(test1, test2, on='bidid')
test_final.head()
```

```
test_final.to_csv('/content/gdrive/My Drive/we_data/FINAL_CSVs/FINAL_TEST_v1')
```
